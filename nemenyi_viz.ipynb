```python
import Orange
import matplotlib.pyplot as plt

avg_ranks = {
    'MISSFOREST': [3.8889, 1.7556, 3.1111, 1.2444],
    'SOFTIMPUTE': [4.0000, 1.8444, 3.0000, 1.1556],
    'GAIN': [4.0000, 1.6444, 2.9333, 1.4222],
    'MIDA': [3.7778, 1.7333, 3.2222, 1.2667],
    'SINKHORN': [4.0000, 1.7778, 2.6000, 1.6222],
    'MICE': [3.7333, 4.5111, 1.9111, 3.3778, 1.4667],
}

names = ['SI', 'MI', 'S-HOT', 'M-HOT']
for method, ranks in avg_ranks.items():
    cd = Orange.evaluation.compute_CD(ranks, 45, alpha='0.05', test='nemenyi')
    print(cd)
    print('Method:', method)
    Orange.evaluation.graph_ranks(ranks, ['MICE'] + names if method == 'MICE' else names,
                                  cd=cd, width=4.5, reverse=True, filename=f'results/nemenyi_{method}.svg')
    plt.show()
```
